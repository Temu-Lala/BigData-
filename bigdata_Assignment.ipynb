```python
dataset = 'languages.csv'
ds = pd.read_csv(dataset)
ds.T
```

```python
ds.dtypes
```

```python
# Get the top 10 languages by users
top_10_users = ds[['title', 'number_of_users']].groupby('title').sum()
top_10_users = top_10_users.nlargest(10, 'number_of_users')

# Get the top 10 languages by jobs
top_10_jobs = ds[['title', 'number_of_jobs']].groupby('title').sum()
top_10_jobs = top_10_jobs.nlargest(10, 'number_of_jobs')

# Create two subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# Plot 1: Top 10 by Users
axs[0].bar(top_10_users.index, top_10_users['number_of_users'], color='skyblue')
axs[0].set_title('Top 10 Languages by Users')
axs[0].set_ylabel('Number of Users')

# Plot 2: Top 10 by Jobs
axs[1].bar(top_10_jobs.index, top_10_jobs['number_of_jobs'], color='lightcoral')
axs[1].set_title('Top 10 Languages by Jobs')
axs[1].set_ylabel('Number of Jobs')


# Rotate x-axis labels for better readability
for ax in axs:
    ax.tick_params(axis='x', rotation=45)

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Graph : Distribution of Languages by Type
plt.figure(figsize=(8, 8))
ds['github_language_type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of Languages by Type')
plt.ylabel('')
plt.show()
```

```python
# Graph 4: Number of Stars vs. Number of Forks
plt.figure(figsize=(10, 6))
sns.scatterplot(x='github_repo_stars', y='github_repo_forks', data=ds)
plt.title('Number of Stars vs. Number of Forks')
plt.xlabel('Number of Stars')
plt.ylabel('Number of Forks')
plt.show()
```

```python
# Graph 5: Distribution of Languages by Type (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(data=ds, x='github_language_type', hue='github_language_type', multiple='stack')
plt.title('Distribution of Languages by Type (Histogram)')
plt.xlabel('Language Type')
plt.ylabel('Number of Repositories')
plt.show()
```

```python
# Graph 6: Number of Users vs. Number of Jobs
plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_users', y='number_of_jobs', data=ds)
plt.title('Number of Users vs. Number of Jobs')
plt.xlabel('Number of Users')
plt.ylabel('Number of Jobs')
plt.show()
```

```python
# Graph 7: Proportion of Open Source vs. Closed Source Languages
plt.figure(figsize=(8, 8))
ds['is_open_source'].value_counts().plot.pie(autopct='%1.1f%%', labels=['Open Source', 'Closed Source'])
plt.title('Proportion of Open Source vs. Closed Source Languages')
plt.ylabel('')
plt.show()
```

```python
# Graph 8: Relationship between Number of Stars and First Commit Date
plt.figure(figsize=(10, 6))
sns.scatterplot(x='github_repo_stars', y='github_repo_first_commit', data=ds)
plt.title('Number of Stars vs. First Commit Date on GitHub')
plt.xlabel('Number of Stars')
plt.ylabel('First Commit Date')
plt.show()
```

```python
# Graph 9: Relationship between Number of Stars and Number of Contributors
plt.figure(figsize=(10, 6))
sns.scatterplot(x='github_repo_stars', y='github_repo_subscribers', data=ds)
plt.title('Number of Stars vs. Number of Subscribers on GitHub')
plt.xlabel('Number of Stars')
plt.ylabel('Number of Subscribers')
plt.show()
```
