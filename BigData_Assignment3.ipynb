# Import libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import dataset


```python
ds = pd.read_csv("Employee.csv")
print(ds)
```

# Data Assessment


```python
ds.info()
```

# Assess dataset duplicate values


```python
duplicate_values = ds.duplicated().sum()
print("Dataset duplicate values:", duplicate_values)
```

# Data Cleaning


```python
clean_dataset = ds.drop_duplicates()
```

# Data Exploration

```python
countplot = sns.countplot(x = "Education", hue = "LeaveOrNot", data = clean_dataset)
for count in countplot.containers:
    countplot.bar_label(count,)
plt.xlabel("Education")
plt.ylabel("Employee Count")
plt.legend(labels = ["stay", "leave"])
plt.title("Employee Turnover by Education", fontweight = "bold")
```
